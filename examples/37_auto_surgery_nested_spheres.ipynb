# Phase 2 Demo: Un-nesting Concentric Spheres in ℝ³

This demo showcases the automated un-nesting of a smaller sphere nested inside a larger sphere in ℝ³.

```python
import sys
import os
sys.path.append(os.path.abspath("../../"))

from tests.discrete_surface_data import build_tetrahedron, to_complex
from pysurgery.autosurgeon import AutoSurgeon
```

## 1. Construct Nested Concentric Spheres

We build two concentric spheres of different radii.

```python
S2_outer = to_complex(build_tetrahedron())
S2_inner = to_complex(build_tetrahedron())
print(f"Outer sphere dimension: {S2_outer.dimension}")
print(f"Inner sphere dimension: {S2_inner.dimension}")
```

## 2. Run the AutoSurgeon

We separate the nested spheres via ambient handle surgery.

```python
surgeon = AutoSurgeon(S2_outer, backend="python")
report = surgeon.run()
print(f"Pipeline status: {report.status}")
print(f"Exact computation: {report.exact}")
print("Plain logs:\n", report.session_logs_plain)
```
