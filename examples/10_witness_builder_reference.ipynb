# 10: Witness Builder Reference

This notebook shows the witness-builder layer for 2D/3D/4D/n>=5, including certificate passthrough keys.

```python
import numpy as np
import scipy.sparse as sp

from pysurgery import (
    ChainComplex,
    FundamentalGroup,
    IntersectionForm,
    ObstructionResult,
    WhiteheadGroup,
    build_homeomorphism_witness,
    build_surface_homeomorphism_witness,
    build_3d_homeomorphism_witness,
    build_4d_homeomorphism_witness,
    build_high_dim_homeomorphism_witness,
)
```

```python
# Surface witness
c2 = ChainComplex(boundaries={1: sp.csr_matrix((1, 0)), 2: sp.csr_matrix((0, 1))}, dimensions=[0, 1, 2])
print(build_surface_homeomorphism_witness(c2, c2).status)
```

```python
# 3D witness with recognition certificate
d1 = sp.csr_matrix(np.zeros((1, 1), dtype=np.int64))
d2 = sp.csr_matrix(np.array([[3]], dtype=np.int64))
d3 = sp.csr_matrix(np.zeros((1, 1), dtype=np.int64))
c3 = ChainComplex(boundaries={1: d1, 2: d2, 3: d3}, dimensions=[0, 1, 2, 3], cells={0: 1, 1: 1, 2: 1, 3: 1})
pi = FundamentalGroup(generators=['g'], relations=[['g', 'g', 'g']])
r3 = build_3d_homeomorphism_witness(c3, c3, pi1_1=pi, pi1_2=pi, recognition_certificate={'provided': True, 'source': 'ref', 'exact': True, 'validated': True})
print(r3.status)
```

```python
# 4D witness with explicit certificate path
q1 = np.array([[1, 0], [0, 1]], dtype=np.int64)
u = np.array([[3, 2], [2, 1]], dtype=np.int64)
q2 = u.T @ q1 @ u
f1 = IntersectionForm(matrix=q1, dimension=4)
f2 = IntersectionForm(matrix=q2, dimension=4)
r4 = build_4d_homeomorphism_witness(f1, f2, ks1=0, ks2=0, simply_connected=True, definite_lattice_isometry_certificate={'provided': True, 'source': 'ref', 'exact': True, 'validated': True, 'isometry_matrix': u.tolist()})
print(r4.status, None if r4.witness is None else r4.witness.kind)
```

```python
# High-dimensional witness with product assembly and completion certificates
d1h = sp.csr_matrix(np.zeros((1, 0), dtype=np.int64))
d5h = sp.csr_matrix(np.zeros((0, 1), dtype=np.int64))
c5 = ChainComplex(boundaries={1: d1h, 5: d5h}, dimensions=[0, 1, 2, 3, 4, 5], cells={0: 1, 1: 0, 2: 0, 3: 0, 4: 0, 5: 1})
wh = WhiteheadGroup(rank=0, description='Wh(Z_2 x Z_3)=0', computable=True, exact=True)
wall = ObstructionResult(dimension=5, pi='Z_2 x Z_3', computable=True, exact=True, value=0, modulus=None, assumptions=[], decomposition_kind='factor_surrogate', assembly_certified=False, obstructs=False, zero_certified=True)
rh = build_high_dim_homeomorphism_witness(c5, c5, dim=5, pi_group='Z_2 x Z_3', whitehead_group=wh, wall_obstruction=wall, homotopy_completion_certificate={'provided': True, 'source': 'ref', 'exact': True, 'validated': True}, product_assembly_certificate={'provided': True, 'source': 'ref', 'exact': True, 'validated': True})
print(rh.status, None if rh.witness is None else rh.witness.kind)
```
