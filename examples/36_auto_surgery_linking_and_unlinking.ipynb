# Phase 1 Demo: Unlinking Hopf-Linked Circles in ℝ³

This demo showcases the automated unlinking of two 1-circles Hopf-linked in ℝ³ using the topology-preserving cancelling-pair surgery flow.

```python
import sys
import os
sys.path.append(os.path.abspath("../../"))

from tests.discrete_surface_data import build_torus, to_complex
from pysurgery.autosurgeon import AutoSurgeon
from pysurgery.auto_surgery import detect_components_with_status
```

## 1. Construct Hopf-Linked Circles

We can build or load the linked circles in the ambient space.

```python
# Load or construct the complex K
torus = to_complex(build_torus())
print(f"Complex dimension: {torus.dimension}")
infos = detect_components_with_status(torus)
print(f"Found {len(infos)} components")
```

## 2. Run the AutoSurgeon

We guide the complex through the unlinking pipeline.

```python
surgeon = AutoSurgeon(torus, backend="python")
report = surgeon.run()
print(f"Pipeline status: {report.status}")
print(f"Exact computation: {report.exact}")
print("Plain logs:\n", report.session_logs_plain)
```
