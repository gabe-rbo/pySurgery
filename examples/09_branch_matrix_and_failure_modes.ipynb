# 09: Branch Matrix and Failure Modes

This notebook demonstrates the major decision outcomes across analyzers:
- success
- impediment
- inconclusive
- surgery_required


```python
import numpy as np
import scipy.sparse as sp

from pysurgery import (
    ChainComplex,
    FundamentalGroup,
    IntersectionForm,
    ObstructionResult,
    WhiteheadGroup,
    analyze_homeomorphism_2d_result,
    analyze_homeomorphism_3d_result,
    analyze_homeomorphism_4d_result,
    analyze_homeomorphism_high_dim_result,
)
```

```python
# 2D success
c2 = ChainComplex(boundaries={1: sp.csr_matrix((1, 0)), 2: sp.csr_matrix((0, 1))}, dimensions=[0, 1, 2])
print('2D:', analyze_homeomorphism_2d_result(c2, c2).status)
```

```python
# 3D inconclusive without decision-ready recognition cert
d1 = sp.csr_matrix(np.zeros((1, 1), dtype=np.int64))
d2 = sp.csr_matrix(np.array([[3]], dtype=np.int64))
d3 = sp.csr_matrix(np.zeros((1, 1), dtype=np.int64))
c3 = ChainComplex(boundaries={1: d1, 2: d2, 3: d3}, dimensions=[0, 1, 2, 3], cells={0: 1, 1: 1, 2: 1, 3: 1})
pi = FundamentalGroup(generators=['g'], relations=[['g', 'g', 'g']])
print('3D:', analyze_homeomorphism_3d_result(c3, c3, pi1_1=pi, pi1_2=pi).status)
```

```python
# 4D impediment by signature mismatch
q1 = IntersectionForm(matrix=np.array([[1]]), dimension=4)
q2 = IntersectionForm(matrix=np.array([[-1]]), dimension=4)
print('4D:', analyze_homeomorphism_4d_result(q1, q2, ks1=0, ks2=0, simply_connected=True).status)
```

```python
# n>=5 surgery_required by non-zero Wall obstruction
d1h = sp.csr_matrix(np.zeros((1, 0), dtype=np.int64))
d5h = sp.csr_matrix(np.zeros((0, 1), dtype=np.int64))
c5 = ChainComplex(boundaries={1: d1h, 5: d5h}, dimensions=[0, 1, 2, 3, 4, 5], cells={0: 1, 1: 0, 2: 0, 3: 0, 4: 0, 5: 1})
wh = WhiteheadGroup(rank=0, description='Wh(1)=0', computable=True, exact=True)
wall = ObstructionResult(dimension=5, pi='1', computable=True, exact=True, value=1, modulus=None, assumptions=[])
print('high-dim:', analyze_homeomorphism_high_dim_result(c5, c5, dim=5, pi1=FundamentalGroup(generators=[], relations=[]), whitehead_group=wh, wall_obstruction=wall).status)
```
